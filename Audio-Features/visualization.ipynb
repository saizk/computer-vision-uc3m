### Visualization of the waves

```python
root_dir = 'full_data'
X, y = load_data(f'{root_dir}/cat', f'{root_dir}/dog')
cats, dogs = X[:164], X[164:]

def plot_waves(n_rows, n_cols, waves):
    f, axs = plt.subplots(n_rows, n_cols, figsize=(14, 8))
    for i in range(n_rows*n_cols):
        plt.subplot(n_rows, n_cols, i+1)
        lbsdisplay.waveplot(waves[i], sr=16000)

    plt.figure(figsize=(12, 4))
    plt.show()
```

#### CATS

```python
plot_waves(2, 2, cats)
```

#### DOGS

```python
plot_waves(2, 2, dogs)
```

```python
def plot_spectrogram(n_rows, n_cols, data):
    f, axs = plt.subplots(n_rows, n_cols, figsize=(14, 8))
    for i in range(n_rows*n_cols):
        plt.subplot(n_rows, n_cols, i+1)
        lbsdisplay.specshow(af.get_mfccs(data[i], sr=16000), sr=16000)

    plt.figure(figsize=(12, 4))
    plt.show()
plot_spectrogram(1, 1, dogs)
```
